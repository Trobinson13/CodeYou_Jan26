```python
BASE_URL = "https://api.weatherapi.com/v1"
```

```python
import os
import requests
import http.client
from dotenv import load_dotenv
```

```python
load_dotenv
WEATHER_API_KEY = os.getenv("API_KEY")
print(WEATHER_API_KEY)
```

```python
params = {
    "key" : WEATHER_API_KEY,
    "q" : "Louisville, KY"
}

response = requests.get(f"{BASE_URL}/current.json", params=params)
print(response)
```

```python
data = response.json()
data
```

```python
def get_current_weather(city: str) -> dict:
    """Fetch current conditions for a city. Returns the parsed JSON dict."""
    response = requests.get(
        f"{BASE_URL}/current.json",
        params={"key": WEATHER_API_KEY, "q": city, "aqi": "no"}
    )
    response.raise_for_status()  # raises an error for 4xx / 5xx status codes
    return response.json()


def get_forecast(city: str, days: int = 3) -> dict:
    """Fetch a multi-day hourly forecast. days must be 1-10 on the free tier."""
    response = requests.get(
        f"{BASE_URL}/forecast.json",
        params={"key": WEATHER_API_KEY, "q": city, "days": days, "aqi": "no", "alerts": "no"}
    )
    response.raise_for_status()
    return response.json()
```

```python
indy_weather = get_current_weather("Indianapolis")
indy_weather
```
