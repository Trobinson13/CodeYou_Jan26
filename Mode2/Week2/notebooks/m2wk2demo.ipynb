## Step 0: Load the Data

Let's start by downloading & reading the CSV.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/jrartd/data-analytics/master/posts-100.csv"
df = pd.read_csv(url)
df.to_csv("../data/posts-100.csv", index=False)
```

```python
print(f"Shape: {df.shape}")
df.head()
```

```python
# Reload with header=None so pandas doesn't eat the first row
df = pd.read_csv(url, header=None)

print(f"Shape: {df.shape}")
df.head()
```

```python
column_names = [
    'post_id', 'post_type', 'creation_date', 'score',
    'view_count', 'last_activity_date', 'title', 'tags',
    'answer_count', 'comment_count', 'favorite_count', 'last_edit_date'
]

df.columns = column_names

df.head()
```
