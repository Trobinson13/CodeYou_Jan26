```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
 
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
avg_streams = spotify_df.groupby("Genre")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_streams
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.xlabel("Genere")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)
plt.xlabel("Genere")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)
plt.xlabel("Genere")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="center") #test between right, centre, and left

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Genere")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="center") #test between right, centre, and left

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values, color = "green", width=0.8)
```

```python
plt.figure(figsize=(10,6))
colors = ["#7bb9c9"]*len(avg_streams)

max_index = avg_streams.idxmax()
highlight_position = avg_streams.index.get_loc(max_index)
colors[highlight_position] = "#7bc99b"

plt.bar(avg_streams.index, avg_streams.values, color = colors, width=0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Genere")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="center") #test between right, centre, and left

plt.tight_layout()
plt.show()
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_artist_streams

plt.figure(figsize=(10,6))
plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)


special_artist = "Ariana Grande"
colors = ["#c97bbc" if artist == special_artist else "#7bb9c9" for artist in avg_artist_streams.index]

max_index = avg_artist_streams.idxmax()
highlight_position = avg_artist_streams.index.get_loc(max_index)
colors[highlight_position] = "#7bc99b"

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45, ha="center") #test between right, centre, and left

plt.tight_layout()
plt.show()
```

```python
df.columns
df_corr = df[['popularity',
       'duration_ms', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']].corr()
df_corr
```

```python
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(df_corr)
```

```python
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(df_corr, annot=True, cmap='mako')
```
