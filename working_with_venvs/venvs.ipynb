# Virtual Environments for Python Projects
Every project deserves its **own little workspace** so its tools and package versions do not collide with anything else on your machine.

That is what a virtual environment does for us. It gives one project its own Python setup, which keeps our work cleaner, safer, and much easier to troubleshoot.


## Step 1: Create and activate your virtual environment

**Do this from the terminal, not inside the notebook.**

Open a terminal in your project's root folder and run:

```bash
python -m venv .venv
```
### Before Running `python -m venv .venv`


### After Running `python -m venv .venv`
We can see the `.venv` dir has been created in the Soultion Explorer view of VS code at the top left and in the Terminals `ls` command


Activate the virtual environment before installing anything:

**Windows PowerShell**
```powershell
.\.venv\Scripts\Activate.ps1
```

**Mac / Linux**
```bash
source .venv/bin/activate
```
Here we can see in the terminal where the session now has the created virtual environment activated

After activation, install packages into the environment you just created:
```bash
python -m pip install ipykernel matplotlib
```

Why do we use `python -m pip` instead of just `pip`?
Because it makes sure the packages are installed by the **same Python interpreter** we are currently using.


## Step 2: Point VS Code at the right interpreter

### In the notebook, make sure the **kernel in the top-right corner** matches that same virtual environment.
Select Kernel -> Choose Your newly created Env<br>
### If you don't see your env in the list
Select Kernel -> Select Another Kernel -> Create Python Environment -> Enter Interpreter Path -> Manually find and select the path of your created venv


## Step 3: Verify the notebook is using the virtual environment

The next code cell uses Python, so it works correctly inside a Python notebook. We are checking:

- Which Python executable is running
- Which Python version the notebook sees
- What the current working directory is
- Whether the environment path appears to be `.venv`


````python
from pathlib import Path
import os
import sys

python_path = Path(sys.executable)
cwd = Path.cwd()

print(f"Python executable: {python_path}")
print(f"Python version: {sys.version.split()[0]}")
print(f"Current working directory: {cwd}")
print(f"Virtual environment active: {'.venv' in str(python_path).lower()}")

if os.environ.get('VIRTUAL_ENV'):
    print(f"VIRTUAL_ENV: {os.environ['VIRTUAL_ENV']}")
else:
    print("VIRTUAL_ENV is not set in this notebook session.")
````
